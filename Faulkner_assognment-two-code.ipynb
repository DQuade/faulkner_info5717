```python
import math
radius = [i for i in range (1, 11)]
ans= []

def volum(r):
	for x in r:
		x= (x ** 3) * math.pi
		z = 4/3 * x
		ans.append(z)
	return ans

sphere = volum(radius)

print ('r', '\t\t\t',  'v') 

for n,g in zip(radius,sphere):

	 print (n,  '\t\t\t',  g)
```
